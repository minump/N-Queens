```python
"""
Assignment 2 NQueens
Question 3 a. Number of assignments made for 8 queens = 168
                Number of assignments made for 10 queens = 179
                Number of assignments made for 15 queens = 1770
                Number of assignments made for 20 queens = 254897
Question 3 b. 100 queens problem is not solvable in a reasonable amount of time.
Question 3 c. Board size of 20 was solvable under a minute. Number of assignments made = 254897
"""

import sys
from queue import *
from collections import *

BOARD_SIZE = int(input("Enter board size: "))
variables=list(range(BOARD_SIZE)) # list of variables
domain=[] # domain of variables
constraints=[] # list of constraints of variables
stack=[] # for storing all assignments made
for var in variables: # initialize domains for each variable
    domain.append(list(range(BOARD_SIZE)))
neighbours={var:list(range(BOARD_SIZE)) for var in variables} # dictionary of varibles to variables
for k,v in neighbours.items(): # to find all arcs/constraints for a variable
    if k in v:
        v.remove(k)
        constraints.append(v)

def AC3(queens, domain, variables, constraints, n):
    q=Queue()# intializing queue for arc consistency
    q=deque([(x,y) for x in variables for y in variables if x!=y])
    while q:
        arc=q.popleft()
        if remove_values(arc,queens,n):
            for i in constraints[arc[0]]:
                q.append((arc[0],i))
    return True # q is empty - all constraints are satisfied
                
def remove_values(arc,queens,n):
    removed=False
    cq=conflicts(arc[0],arc[1],queens)
    if n in cq:
        ind=cq.index(n) # find domain value where number of conflicts is n
        #print("remove ",ind, "from domain of ",arc[0], domain[arc[0]])
        domain[arc[0]].remove(ind)
        removed=True
    return removed


def conflicts(var,row,queens): # Counts the number of attacks 
    conflictqueens=[0]*BOARD_SIZE # list for storing the number of conflicts for each queen in row
    for i in domain[var]:
        count=0
        for j in domain[row]:
            if under_attack(j,queens+[i]):
                count+=1
        conflictqueens[i]=count
    return conflictqueens

def under_attack(col, queens):
    return col in queens or \
           any(abs(col - x) == len(queens)-i for i,x in enumerate(queens))
 
def rsolve(queens,n):
    if n == len(queens):
        return queens
    else:
        if AC3(queens,domain, variables, constraints, n):
            for i in range(n):
                for value in domain[i]:
                    if not under_attack(value,queens):
                        stack.append(value)
                        newqueens = rsolve(queens+[value],n)
                        if newqueens != []:
                            return newqueens
        return [] # FAIL

    
def print_board(queens):
    row = 0
    n = len(queens)
    for pos in queens:
        for i in range(pos):
            sys.stdout.write( ". ")
        sys.stdout.write( "Q ")
        for i in range((n-pos)-1):
            sys.stdout.write( ". ")
        print('\n')
    

ans = rsolve([],BOARD_SIZE)
print_board(ans)
#print(stack) # Print all assignments made
print("Total number of assignments: ",len(stack))
```
