```python
"""
Assignment 2 NQueens
Question 1 a. Number of assignments made for 8 queens = 113
                Number of assignments made for 10 queens = 102
                Number of assignments made for 15 queens = 1359
                Number of assignments made for 20 queens = 199635
Question 1 b. 100 queens problem is not solvable in a reasonable amount of time.
"""

import sys

BOARD_SIZE = int(input("Enter board size: "))
stack=[] # For storing all assignments made


def under_attack(col, queens):
    return col in queens or \
           any(abs(col - x) == len(queens)-i for i,x in enumerate(queens))
 
def rsolve(queens,n):
    if n == len(queens):
        return queens
    else:
        for i in range(n):
            if not under_attack(i,queens):
                stack.append(i)
                newqueens = rsolve(queens+[i],n)
                if newqueens != []:
                    return newqueens
        return [] # FAIL

def print_board(queens):
    row = 0
    n = len(queens)
    for pos in queens:
        for i in range(pos):
            sys.stdout.write( ". ")
        sys.stdout.write( "Q ")
        for i in range((n-pos)-1):
            sys.stdout.write( ". ")
        print('\n')
    

ans = rsolve([],BOARD_SIZE)
print_board(ans)
#print(stack) # Print all assignments made
print('\n')
print("Total number of assignments: ",len(stack))
```
